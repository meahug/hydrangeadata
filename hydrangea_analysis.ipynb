```python
# Here's my notebook for hydrangea location, color, approximate pH, and whether each hydrangea was potted or not.
# I used geojson to plot the lat/long, which I approximated based on google maps screenshots as I was walking and recording + taking photos of hydrangeas.
# Then I downloaded as .csv from geojson and brought this into jupyter notebooks

import pandas as pd
df = pd.read_csv('hydrangeapoints.csv')
```

```python
# taking the overall mean pH

df['pH'].mean()
```
